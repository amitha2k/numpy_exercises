Q7. Lex x be an array <br/>
[[ 1 2 3]<br/>
[ 4 5 6].<br/><br/>
and y be an array <br/>
[[ 7 8 9]<br/>
[10 11 12]].<br/>
Concatenate x and y so that a new array looks like <br/>[[1, 2, 3, 7, 8, 9], <br/>[4, 5, 6, 10, 11, 12]].


```python
'''
[[ 1  2  3  7  8  9]
 [ 4  5  6 10 11 12]]
'''
x = np.array([[1,2,3],[4,5,6]])
y = np.array([[7,8,9],[10,11,12]])
np.vstack([np.hstack([x[0],y[0]]),np.hstack([x[1],y[1]])])
```

Q14. Let x be an array [0, 1, 2]. Convert it to <br/>
[[0, 1, 2, 0, 1, 2],<br/>
 [0, 1, 2, 0, 1, 2]].

```python
'''
[[0 1 2 0 1 2]
 [0 1 2 0 1 2]]
'''
x = np.array([0,1,2])
np.vstack([np.concatenate((x,x),axis=0),np.concatenate((x,x),axis=0)])
```
